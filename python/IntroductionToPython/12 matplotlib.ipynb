```python
city = ['', '', '', '', '', '', '']

# (latitude) (longitude)
lat  = [37.56, 37.45, 36.35, 35.87, 35.53, 35.18, 35.16] 
lon = [126.97, 126.70, 127.38, 128.60, 129.31, 129.07, 126.85]

#  (/km^2): 2017  
pop_den = [16154, 2751, 2839, 2790, 1099, 4454, 2995]

size = np.array(pop_den) * 0.2 #    
colors = ['r', 'g', 'b', 'c', 'm', 'k', 'y'] #   

plt.scatter(lon, lat, s=size, c=colors, alpha=0.5)
plt.xlabel('(longitude)')
plt.ylabel('(latitude)')
plt.title('  (2017)')

for x, y, name in zip(lon, lat, city):
    plt.text(x, y, name) #      

plt.show()
```

```python
member_IDs = ['m_01', 'm_02', 'm_03', 'm_04']
before_ex = [27, 35, 40, 33]
after_ex = [30, 38, 42, 37] 
n_data = len(member_IDs)
index = np.arange(n_data)

colors=['r', 'g', 'b', 'm']
plt.bar(index, before_ex, color = colors, alpha = 0.4, tick_label = member_IDs, width = 0.6)
plt.show()
```

```python
colors=['r', 'g', 'b', 'm']
plt.barh(index, before_ex, color = colors, alpha = 0.4, tick_label = member_IDs)
plt.show()
```

```python
barWidth=0.4
plt.bar(index, before_ex, color='c', alpha=0.4, align='edge', width=barWidth, label='before')
plt.bar(index+barWidth, after_ex, color='m', alpha=0.4, align='edge', width=barWidth, label='after')

plt.xticks(index + barWidth, member_IDs)
plt.legend()
plt.xlabel(' ID')
plt.ylabel(' ')
plt.title('    ()  ')
plt.show()
```
