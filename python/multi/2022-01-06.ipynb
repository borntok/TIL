### 큐의 간단 구현

```python
def isQueueFull():
    global size, queue, front, rear
    if rear == (size-1):
        return True
    else:
        return False
    
def enQueue(data):
    global size, queue, front, rear
    if isQueueFull():
        print('풀방')
        return
    rear += 1
    queue[rear] = data
```

```python
def isQueueEmpty():
    global size, queue, front, rear
    if front == rear:
        return True
    else:
        return False
    
def deQueue():
    global size, queue, front, rear
    if isQueueEmpty():
        print('빈 큐')
        return None
    front += 1
    data = queue[front]
    queue[front] = None
    return data

def peek():
    global size, queue, front, rear
    if isQueueEmpty():
        print('빈 큐')
        return None
    return queue[front+1]
```

### 큐의 일반 구현

```python
def isQueueFull():
    global size, queue, front, rear
    if rear != size-1:
        return False
    elif (rear == size-1) and (front == -1):
        return True
    else:
        for i in range(front+1, size):
            queue[i-1] = queue[i]
            queue[i] = None
        front -= 1
        rear -= 1
        return False
```

```python
SIZE = 5
queue = [None for _ in range(SIZE)]
front = rear = -1

enQueue('화사')
enQueue('솔라')
enQueue('문별')
enQueue('휘인')
enQueue('선미')
print('출구<--', queue, '<--입구')
print('입장 손님 : ', deQueue())
print('입장 손님 : ', deQueue())
print('출구<--', queue, '<--입구')
enQueue('재남')
print('출구<--', queue, '<--입구')
enQueue('BTS')
print('출구<--', queue, '<--입구')
enQueue('울랄라')
print('출구<--', queue, '<--입구')
```

### 원형큐
- rear+1과 front가 같은 경우 원형 큐가 꽉 찬 것으로 처리

```python
def isQueueFull():
    global size, queue, front, rear
    if (rear+1) % size == front:
        return True
    else:
        return False
    
def enQueue(data):
    global size, queue, front, rear
    if isQueueFull():
        print('풀방')
        return
    rear = (rear+1) % size
    queue[rear] = data

def isQueueEmpty():
    global size, queue, front, rear
    if front == rear:
        return True
    else:
        return False
    
def deQueue():
    global size, queue, front, rear
    if isQueueEmpty():
        print('빈 큐')
        return None
    front = (front+1) % size
    data = queue[front]
    queue[front] = None
    return data

def peek():
    global size, queue, front, rear
    if isQueueEmpty():
        print('빈 큐')
        return None
    return queue[(front+1) % size]    
```

```python
SIZE = 5
queue = [None for _ in range(SIZE)]
front = rear = -1

enQueue('화사')
enQueue('솔라')
enQueue('문별')
enQueue('휘인')
enQueue('선미')
print('출구<--', queue, '<--입구')
print('입장 손님 : ', deQueue())
print('입장 손님 : ', deQueue())
print('출구<--', queue, '<--입구')
enQueue('선미')
print('출구<--', queue, '<--입구')
```

### 이진탐색트리 일반 구현

```python
memory = []
root = None
nameAry = ['블랙핑크', '레드벨벳', '마마무', '에이핑크', '걸스데이', '트와이스']
```

```python
node = TreeNode()
node.data = nameAry[0]
root = node
memory.append(node)

for name in nameAry[1:]:
    node = TreeNode()
    node.data = name
    
    current = root
    while True:
        if name < current.data :
            if current.left == None:
                current.left = node
                break
            current = current.left
        else:
            if current.right == None:
                current.right = node
                break
            current = current.right
            
    memory.append(node)
    
print('이진탐색트리 구성 완료')
```

```python
findName = '마마무'
current = root
while True :
    if findName == current.data:
        print(findName, '찾음!')
        break
    elif findName < current.data:
        if current.left == None :
            print(findName, '없음')
            break
        current = current.left
    else:
        if current.right == None :
            print(findName, '없음')
            break
        current = current.right
        
print('종료')
```

## 정렬

### 선택 정렬

```python
def findMinIndex(ary):
    minIdx = 0
    for i in range(1, len(ary)):
        if ary[minIdx] > ary[i]:
            minIdx = i
    return minIdx
```

```python
# testAry = [55, 88, 33, 77]
import random
testAry = [random.randint(1, 99) for _ in range(20)]
```

```python
print(testAry)
minPos = findMinIndex(testAry)
print('최솟값-->', testAry[minPos])
```

```python
before = [random.randint(10, 99) for _ in range(20)]
after = []
```

```python
print('정렬 전-->', before)
for i in range(len(before)):
    minPos = findMinIndex(before)
    after.append(before[minPos])
    del before[minPos]
print('정렬 후-->', after)
```

### 개선된 선택 정렬

```python
dataAry = [random.randint(10, 99) for _ in range(20)]
```

```python
def selectionSort(ary):
    n = len(ary)
    for cy in range(0, n-1):
        minIdx = cy
        for i in range(cy+1, len(ary)):
            if ary[minIdx] > ary[i]:
                minIdx = i
        ary[cy], ary[minIdx] = ary[minIdx], ary[cy]

        return ary
```

```python
print('정렬 전-->', dataAry)
dataAry = selectionSort(dataAry)
print('정렬 후-->', dataAry)
```

## 검색

### 순차 검색

```python
dataAry = [random.randint(10, 99) for _ in range(20)]
findData = dataAry[random.randint(0, 19)]
```

```python
def seqSearch(ary, fData):
    pos = -1
    for i in range(len(ary)):
        if ary[i] == fData:
            pos = i
            break
    return pos
```

```python
print('배열-->', dataAry)
position = seqSearch(dataAry, findData)
if position == -1:
    print(findData, '없다')
else:
    print(findData, '는', position, '위치에 있어요')
```

### 이진 검색

```python
dataAry = [random.randint(1, 999) for _ in range(10)]
findData = dataAry[random.randint(0, 19)]
dataAry.sort()
```

```python
def binSearch(ary, fData):
    pos = -1
    start = 0
    end = len(ary)-1
    
    while start <= end:
        mid = (start + end) // 2
        if fData == ary[mid]:
            return mid
        elif fData > ary[mid]:
            start = mid + 1
        else:
            end = mid - 1
    
    return pos
```

```python
print('배열-->', dataAry)
position = seqSearch(dataAry, findData)
if position == -1:
    print(findData, '없다')
else:
    print(findData, '는', position, '위치에 있어요')
```
