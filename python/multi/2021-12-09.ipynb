```python
# isin은 리스트를  파라미터로 갖음
# 리스트 네에 하나라도 존재하면 True 그렇지 않으면 False

rawData[ (rawData['지점'] == '서초구')&(rawData['담당'].isin(['김현경', '최현경']) )]
```

```python
# apply의 기본적인 사용법

def func(x):
  print('함수가 호출되었습니다')
  print(f'x:{x}')
  return

rawData.apply(func)
```

```python
rawData.apply(func, axis=1) #행별 apply 적용
```

```python
#지점별로 지역을 새로 만들기

def func(x):
  if x['지점'] == '관악구':
    return '강서'
  elif x['지점'] == '여의도':
    return '강서'
  elif x['지점'] == '강남구':
    return '강동'
  elif x['지점'] == '서초구' :
    return '강동'

rawData['지역'] = rawData.apply(func, axis=1)
rawData
```
